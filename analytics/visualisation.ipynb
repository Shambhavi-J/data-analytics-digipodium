```python
import numpy as np
import plotly.express as px
```

```python
xpoints = np.arange(70)
ypoints =  np.random.rand(70)
px.line(x=xpoints, y=ypoints)
```

```python
x = np.random.randint(1, 100, 100)
y = np.random.randint (1, 100,100)
z = np.random.uniform(1, 100, 100)
px.scatter_3d(x=x, y=y, z=z,
               color = z,
               size=z,
               size_max=18,
               template='plotly_dark')
```

```python
tips = px.data.tips()
tips
```

```python
#pie chart
count = tips['sex'].value_counts()
px.pie(count, count.index, count.values)
```

```python
#donutchart
px.pie(count,count.index, count.values, hole=0.5)
```

```python
# compare two numerical data
px.scatter(tips, x='total_bill', y='tip', size='size', template='plotly_dark')
```

```python
px.scatter(tips, x='total_bill',y='tip', facet_col='sex', facet_row='smoker')
```

```python
#sunburst
px.sunburst(tips,path =['day','time','sex'],values='total_bill',template='plotly_dark')
```

```python
iris= px.data.iris()
iris
```

```python
fig= px.scatter_matrix(iris,
                    dimensions=['sepal_width',
                                'sepal_length',
                                'petal_width',
                                'petal_length'
                    ],
                    color='species',
                    height=800, width=1000)
fig.update_layout(title='scatter matrix of iris',
                  dragmode='select',
                  hovermode='closest')
fig.show()
```
